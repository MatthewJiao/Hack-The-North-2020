```python
i=0
to_drop = []


while i<rows-2:
    value1 = df1.iloc[i][3]
    value2 = df1.iloc[i+1][3]
    value3 = df1.iloc[i+2][3]

    if value1 == 0.0 or value1 == 0:
        if value2 == 0.0 or value2 == 0:
            to_drop.append(i)

        elif value3 == 0.0 or value3 == 0:
            to_drop.append(i)
            to_drop.append(i+1)

    i+=1
    
print(len(to_drop))
print(to_drop)
```

```python
df1 = df1.drop(to_drop)
df1
```
