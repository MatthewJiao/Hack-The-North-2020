```python
i=0
most = []
pre = []
post = []
secondary = []
while i<rows:
   # print(df1.iloc[i][12])
    if df1.iloc[i][12] == "M":
            most.append(1)
            pre.append(0)
            post.append(0)
            secondary.append(0)
    elif df1.iloc[i][12] == "1":
            most.append(0)
            pre.append(1)
            post.append(0)
            secondary.append(0)
    elif df1.iloc[i][12] == "2":
            most.append(0)
            pre.append(0)
            post.append(1)
            secondary.append(0)
    elif df1.iloc[i][12] == "3":
            most.append(0)
            pre.append(0)
            post.append(0)
            secondary.append(1)
    elif math.isnan(df1.iloc[i][12]):
            most.append(0)
            pre.append(0)
            post.append(0)
            secondary.append(0)
    i+=1
    
```

```python
del df1["Diag_type"]
df1
```

```python
df1["Most_responsible"] = most
df1["pre_admit"] = pre
df1["post_admit"] = post
df1["secondary"] = secondary
```

```python
df1 = df1[['pt_identifier', 'new_date', 'BP1', 'BP2', 'Pulse', 'Temp', 'RR', 'O2Sat',
       'Age_Number', 'Gender', 'Weight', 'Diag_desc',
       'Most_responsible', 'pre_admit', 'post_admit', 'secondary', 'Lactate_Value', 'Lactate_High_Low',
       'Lactate_Ref_Range', 'WBC_Value', 'WBC_High_Low', 'WBC_Ref_Range']]
```
